Imports

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

1) Carregar o dataset

```python
dados = load_iris()
X = dados.data          # Features (medidas)
y = dados.target        # Labels (espécies: 0,1,2)
```

2) Separar treino e teste

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=50135
)
```

3) Criar e treinar o modelo

```python
modelo = DecisionTreeClassifier(random_state=50135)
modelo.fit(X_train, y_train)
```

4) Prever e valiar

```python
y_pred_ml = modelo.predict(X_test)
acc_ml = accuracy_score(y_test, y_pred_ml)
```

5) Print do modelo

```python
print("=== ML (Decision Tree) ===")
print("Acurácia:", round(acc_ml, 4))
print("Exemplo de previsões:", y_pred_ml[:10])
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred_ml)

nomes_classes = dados.target_names

fig, ax = plt.subplots(figsize=(6, 6))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=nomes_classes)

disp.plot(cmap='Blues', ax=ax, values_format='d')

plt.title('Matriz de Confusão - Árvore de Decisão', fontsize=14, pad=15)
plt.xlabel('Predição do Modelo', fontsize=11)
plt.ylabel('Classe Real', fontsize=11)
plt.show()
```
